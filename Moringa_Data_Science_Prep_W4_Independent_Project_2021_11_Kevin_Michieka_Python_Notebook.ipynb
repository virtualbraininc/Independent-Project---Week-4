```python
# import python libraries
import pandas as pd
import numpy as np
from datetime import date, time, datetime
```

```python
# Read and preview the CSV dataset first 10 rows
Electric_Cars = pd.read_csv('http://bit.ly/autolib_dataset')
Electric_Cars.head(10)
```

```python
# Preview the last 5 rows of the dataframe
Electric_Cars.tail(5)
```

```python
# Investigate genreal information about out dataset
Electric_Cars.info()
```

```python
# Data Cleaning
#Start by removing unncesesary information by dropping some columns.
Electric_Cars.drop(['Displayed comment', 'ID', 'Geo point','Address', 'Cars'], axis = 1, inplace = True)
Electric_Cars
```

```python
Electric_Cars.describe()
```

```python
# Identifying Outliers with Interquartile Range (IQR)
Q1 = Electric_Cars.quantile(0.25)
Q3 = Electric_Cars.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# The above output prints the IQR scores, which can be used to detect outliers. 
#The code below generates an output with the 'True' and 'False' values.
# Points where the values are 'True' represent the presence of the outlier
(Electric_Cars < (Q1 - 1.5 * IQR)) |(Electric_Cars > (Q3 + 1.5 * IQR))
```

```python
# Above results shows that the data is accurate
# Lets check for completeness
print(Electric_Cars.isnull().any())
```

```python
# Count total NaN in a DataFrame
print(" \nCount total NaN in a DataFrame : \n\n",
       Electric_Cars.isnull().sum().sum())
```

```python
# Checking consistency and remaving any deplicates.
Electric_Cars[Electric_Cars.duplicated()]
# drop duplicates
Electric_Cars_2 = Electric_Cars.drop_duplicates()
# check for non duplicates
Electric_Cars_2.head(10)
```

```python
# Car movement can be determined by the change is the number of cars at the station anginest the numbers at counters.
Electric_Cars_2['Bluecar_In/Out'] = Electric_Cars_2['Bluecar counter'].diff()
Electric_Cars_2.head(10)
```

**Research Question**

Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
#Focus on car movements in paris by creating a paris dataframe

paris_Bluecar = Electric_Cars_2[Electric_Cars_2['City'] =='Paris'] 

paris_Bluecar.head(5)
```

```python
# Answer
paris_Bluecar[paris_Bluecar['Bluecar_In/Out'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```
